```python
# Tokenization, Lemmatization
# Data Cleansing: Remove stopwords, remove symbols, remove URLS.
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string


def preprocess_text(text):
    # Tokenize the text
    tokens = word_tokenize(text)

    # Convert tokens to lowercase
    tokens = [token.lower() for token in tokens]

    # Remove punctuation symbols
    tokens = [token for token in tokens if token not in string.punctuation]

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    # Join the tokens back into a clean text
    clean_text = ' '.join(tokens)

    return clean_text

# Apply preprocessing to the 'text_column'
df['cleaned_text'] = df['Comments'].apply(preprocess_text)

df
```

```python
    # a) CountVectorizer
    # b) TFIDFVectorizer
    # c) Word2Vec
    # d) GoogleNews Word2Vec
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from gensim.models import Word2Vec, KeyedVectors
import string

# Sample DataFrame with 'Comments' column
data = {'Comments': ["This is the first comment.",
                     "This comment is the second comment.",
                     "And this is the third one.",
                     "Is this the first comment?"]}
df = pd.DataFrame(data)

# Tokenize function for Word2Vec
def tokenize_and_lowercase(text):
  return word_tokenize(text.lower())

# Apply tokenization to the 'Comments' column
df['Tokenized_Comments'] = df['Comments'].apply(tokenize_and_lowercase)

# Convert the tokenized comments back to strings for CountVectorizer and TF-IDF
df['Tokenized_Comments_String'] = df['Tokenized_Comments'].apply(' '.join)

# a) CountVectorizer
vectorizer_count = CountVectorizer()
X_count = vectorizer_count.fit_transform(df['Tokenized_Comments_String'])
df_count = pd.DataFrame(X_count.toarray(), columns=vectorizer_count.get_feature_names_out())
print("CountVectorizer:\n", df_count)

# b) TF-IDF Vectorizer
vectorizer_tfidf = TfidfVectorizer()
X_tfidf = vectorizer_tfidf.fit_transform(df['Tokenized_Comments_String'])
df_tfidf = pd.DataFrame(X_tfidf.toarray(), columns=vectorizer_tfidf.get_feature_names_out())
print("\nTF-IDF Vectorizer:\n", df_tfidf)

# c) Word2Vec
model_w2v = Word2Vec(sentences=df['Tokenized_Comments'], vector_size=100, window=5, min_count=1, sg=0)
word_vectors = {word: model_w2v.wv[word] for word in model_w2v.wv.index_to_key}
print("\nWord2Vec Vector for 'comment':", word_vectors['comment'])

# d) Google News Word2Vec (pre-trained)
path_to_google_w2v = "/content/drive/MyDrive/GoogleNews-vectors-negative300.bin"
model_google_w2v = KeyedVectors.load_word2vec_format(path_to_google_w2v, binary=True)
print("\nGoogle News Word2Vec Vector for 'king':", model_google_w2v['king'])
```
